## 3. 涨跌停制度

### 3.1 涨跌停幅度

| 板块 | 涨跌停幅度 | 特殊情况 |
|------|------------|----------|
| 主板/中小板 | ±10% | ST 股票 ±5% |
| 创业板/科创板 | ±20% | 上市首日无限制 |
| 北交所 | ±30% | 上市首日无限制 |
| 新股首日 | ±44%（沪） | - |

### 3.2 涨跌停计算规则

- 涨跌停价 = 前收盘价 × (1 ± 涨跌停幅度)
- 涨停价：当天最高可以挂单的价格
- 跌停价：当天最低可以挂单的价格
- **一字涨停/跌停**：开盘即涨停/跌停，全天封板无法买入/卖出

## 4. T+1 交易制度

> T+1 = Trade + 1 Day

**当天买入的股票，当天不能卖出，必须等到下一个交易日才能卖出。**

**当天卖出股票的资金，可以马上用于买入，但当天不能转出到银行卡。**

```python
# 示例说明
T日（今天）: 买入股票 A
T+1日（明天）: 可以卖出股票 A
```

### T+1 的影响
1. **当日买入当日卖出不可能**：避免了日内频繁交易
2. **资金锁定**：卖出后资金 T+1 可用
3. **增加了隔夜风险**：持仓过夜面临夜间不确定性

## 5. 手续费结构

### 5.1 手续费组成

| 费用项 | 收费标准 | 收取方式 |
|--------|----------|----------|
| 印花税 | 成交金额的 0.05%（卖出时） | 国家强制收取 |
| 过户费 | 成交金额的 0.001%（双向） | 中国结算收取 |
| 佣金（手续费） | 不超过成交金额 0.3%（双向） | 券商收取 |
| 规费 | 约 0.002%（双向） | 交易所收取 |

### 5.2 实际计算示例

```python
# 手续费计算示例
def calculate_fee(price, quantity, side='buy'):
    amount = price * quantity
    stamp_tax = amount * 0.0005 if side == 'sell' else 0
    transfer_fee = amount * 0.00001
    commission = max(amount * 0.0003, 5)
    regulation_fee = amount * 0.00002
    total_fee = stamp_tax + transfer_fee + commission + regulation_fee
    return {
        'amount': amount,
        'stamp_tax': stamp_tax,
        'transfer_fee': transfer_fee,
        'commission': commission,
        'regulation_fee': regulation_fee,
        'total_fee': total_fee,
        'cost_ratio': total_fee / amount * 100
    }
```

````python
# 手续费计算示例
def calculate_fee(price, quantity, side='buy'):
    """计算交易手续费"""
    amount = price * quantity  # 成交金额

    # 印花税（仅卖出收取）
    stamp_tax = amount * 0.0005 if side == 'sell' else 0

    # 过户费（双向收取）
    transfer_fee = amount * 0.00001

    # 佣金（双向收取，假设万三，最低 5 元）
    commission = max(amount * 0.0003, 5)

    # 规费（假设万二）
    regulation_fee = amount * 0.00002

    total_fee = stamp_tax + transfer_fee + commission + regulation_fee

    return {
        'amount': amount,
        'stamp_tax': stamp_tax,
        'transfer_fee': transfer_fee,
        'commission': commission,
        'regulation_fee': regulation_fee,
        'total_fee': total_fee,
        'cost_ratio': total_fee / amount * 100  # 成本占比
    }

# 例子：10 元价格买入 1000 股
result = calculate_fee(price=10, quantity=1000, side='buy')
print(f"买入 10元 × 1000股:")
print(f"  成交金额: {result['amount']:.2f} 元")
print(f"  过户费: {result['transfer_fee']:.2f} 元")
print(f"  佣金: {result['commission']:.2f} 元")
print(f"  规费: {result['regulation_fee']:.2f} 元")
print(f"  总费用: {result['total_fee']:.2f} 元")
print(f"  成本占比: {result['cost_ratio']:.3f}%")

# 卖出时多了印花税
result_sell = calculate_fee(price=11, quantity=1000, side='sell')
print(f"\n卖出 11元 × 1000股:")
print(f"  成交金额: {result_sell['amount']:.2f} 元")
print(f"  印花税: {result_sell['stamp_tax']:.2f} 元")
print(f"  总费用: {result_sell['total_fee']:.2f} 元")
print(f"  成本占比: {result_sell['cost_ratio']:.3f}%")
````
